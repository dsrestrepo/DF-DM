```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Data

```python
df = pd.read_csv('Embeddings_c/mimic/text_embeddings.csv')
df
```

### EDA

```python
# Quick EDA with percentages
# Total counts for each category to calculate percentages
total_sex = df["sex_label"].count()
total_race = df["race_label"].count()
total_disease = df["disease_label"].count()

# Calculating percentages
sex_counts["Percentage"] = (sex_counts["Frequency"] / total_sex) * 100
race_counts["Percentage"] = (race_counts["Frequency"] / total_race) * 100
disease_counts["Percentage"] = (disease_counts["Frequency"] / total_disease) * 100

# Displaying updated counts with percentages
sex_counts, race_counts, disease_counts
```

```python
# Quick EDA
# Counts per sex_label
sex_counts = df["sex_label"].value_counts().reset_index()
sex_counts.columns = ['Sex', 'Frequency']

# Counts per race_label
race_counts = df["race_label"].value_counts().reset_index()
race_counts.columns = ['Race', 'Frequency']

# Counts per disease_label
disease_counts = df["disease_label"].value_counts().reset_index()
disease_counts.columns = ['Disease', 'Frequency']

# Plotting
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

sns.barplot(x='Sex', y='Frequency', data=sex_counts, ax=axs[0]).set_title('Sex Distribution')
sns.barplot(x='Race', y='Frequency', data=race_counts, ax=axs[1]).set_title('Race Distribution')
sns.barplot(x='Disease', y='Frequency', data=disease_counts, ax=axs[2]).set_title('Disease Distribution')

plt.tight_layout()
plt.show()

sex_counts, race_counts, disease_counts
```

```python
# Function to plot distributions per split
def plot_distributions_per_split(df, attribute, title):
    splits = df['split'].unique()
    fig, axs = plt.subplots(1, len(splits), figsize=(18, 6), sharey=True)
    for i, split in enumerate(splits):
        split_data = df[df['split'] == split][attribute].value_counts().reset_index()
        print(f'Distribution {split_data}: {split_data}')
        split_data.columns = [attribute, 'Frequency']
        sns.barplot(x=attribute, y='Frequency', data=split_data, ax=axs[i]).set_title(f'{title} in {split.capitalize()} Split')
    plt.tight_layout()
    plt.show()

# Plotting distributions per split for 'sex_label', 'race_label', and 'disease_label'
plot_distributions_per_split(df, 'sex_label', 'Sex Distribution')
plot_distributions_per_split(df, 'race_label', 'Race Distribution')
plot_distributions_per_split(df, 'disease_label', 'Disease Distribution')
```
